# CONVERT VIDEO FRAMES TO IMAGES

#### This notebook is to provide an easy way to save a video to its individual frames to a specified folder

```python
#impoert relevant libraries
import cv2
import os
import matplotlib.pyplot as plt
```

##### UPDATE THE PATHS BELOW TO USE NOTEBOOK

```python

video_path = "../img_training_data/IMG_3024.MOV"
save_to_path = "../img_training_data/raw_images/"
```

```python
#READ IN VIDEO CAPTURE
cap= cv2.VideoCapture(video_path)
ret, frame = cap.read()
```

```python
frame = cv2.flip(src=frame,flipCode=0)
```

```python
#VERIFY THAT FRAME SHOWS UP
plt.imshow(frame)
```

```python
#MAKE SURE THE IMAGE SAVES TO THE RIGHT PLACE
cv2.imwrite(save_to_path +"img_0.png" ,frame)
```

##### RUN WHEN YOU KNOW THE CODE ABOVE WORKS

```python
#RUN THE REST OF THE SCRIPT
i=1
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    frame = cv2.flip(src=frame,flipCode=0)
    cv2.imwrite(save_to_path + f"img_{i}.png", frame)
    i+=1

cap.release()
cv2.destroyAllWindows()
print("DONE")
```
