# Making Image Filter for Dokalman

##### In this notebook, we will filter out a single image in the goal of detecting just the white line. Eventually this notebook may be expanded to many images in a video or from a folder

### Load in Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Read in first image

```python
img = cv2.imread("test_img1.png")
```

```python
plt.imshow(img)
```

### Convert to black/white image

```python
img_bw = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_bw)
```

### Threshold out white line

```python
x, img_thr = cv2.threshold(img, 230,255, cv2.THRESH_BINARY)
```

```python
plt.imshow(img_thr)
```
